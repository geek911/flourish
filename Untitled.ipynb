```python
def age_at_enrollment(child_dob=None, enrollment_date=None):
    """Returns age months as decimals.
    """
    child_dob = child_dob

    child_age = age(child_dob, enrollment_date)

    child_age = str(child_age.years) + '.' + str(child_age.months)
    return float(child_age)
```

```python
def age_at_year_3(child_dob):
    """Returns the age at year 3.
    """
    start_date_year_3 = datetime.datetime(2022, 7, 1).date()
    
    return age_at_enrollment(child_dob=child_dob, enrollment_date=start_date_year_3)
```
