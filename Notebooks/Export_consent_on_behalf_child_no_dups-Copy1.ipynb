```python
# Imports
import os 
import pandas as pd
from django.core.exceptions import ValidationError
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
from datetime import date

data_path = '/home/flourish/'
```

```python
ccc = CaregiverChildConsent.objects.all()
```

```python
child_consent_df = pd.DataFrame([s.__dict__ for s in ccc])
```

```python
child_consent_df
```

```python
child_consent_df1 = child_consent_df.sort_values('consent_datetime').drop_duplicates(
    subset=['subject_identifier'], keep='last')
```

```python
child_consent_df1.count()
```

```python
list(child_consent_df1.columns)
```

```python
child_consent_df1['child_dob'] = pd.to_datetime(child_consent_df1['child_dob']).dt.date
child_consent_df1['child_dob'] = pd.to_datetime(child_consent_df1['child_dob'], format='%Y/%m/%d')


child_consent_df1['is_verified_datetime'] = pd.to_datetime(child_consent_df1['is_verified_datetime']).dt.date
child_consent_df1['is_verified_datetime'] = pd.to_datetime(child_consent_df1['is_verified_datetime'], format='%Y/%m/%d')

child_consent_df1['consent_datetime'] = pd.to_datetime(child_consent_df1['consent_datetime']).dt.date
child_consent_df1['consent_datetime'] = pd.to_datetime(child_consent_df1['consent_datetime'], format='%Y/%m/%d')

child_consent_df1['report_datetime'] = pd.to_datetime(child_consent_df1['report_datetime']).dt.date
child_consent_df1['report_datetime'] = pd.to_datetime(child_consent_df1['report_datetime'], format='%Y/%m/%d')
```

```python
child_consent_df1 = child_consent_df1.drop(['_state', 'first_name', 'last_name', 'identity', 'confirm_identity',
                                            'guardian_name', 'initials', 'revision', 'dob', 'created', 'modified'],
                                           axis=1)
```

```python
# saving the dataframe
child_consent_df1.to_csv(f'{data_path}caregiver_child_consent_202204211807.csv')
```
